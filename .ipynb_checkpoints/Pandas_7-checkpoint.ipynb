```python
taitanic = pd.read_csv("titanic.csv")
taitanic["Age"].mean()
taitanic[["Age","Fare"]].mean()
taitanic[["Age","Fare"]].median()
taitanic[["Age","Fare"]].describe()
taitanic.agg(# how to calculate my statistical summary
    {
        "Age":["min","max","median"],
        "Fare":["min","max","median"]
    }
)


taitanic[["Sex","Age"]].groupby("Sex").mean() # ~ taitanic.groupby("Sex")["Age"].mean()
taitanic.groupby("Embarked")["Fare"].mean()
taitanic.groupby("Sex").mean()


#the number of the passengers in each of the cabin class
print(taitanic["Pclass"].value_counts()) # ~ print(taitanic.groupby("Pclass")["Pclass"].size())


#Size Vs Count
# Size -> Nan = True    #Include Counts Of Nan
# Count -> Nan = False  #Dont include Counts of Nan
#value_counts(drona = True) -> Nan = False

```
