```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "key":["k0","k1","k2","k3","k4","k5"],
    "A":["A0","A1","A2","A3","A4","A5"]
})
df2 = pd.DataFrame({
    "key":["k0","k1","k2"],
    "B":["B0","B1","B2"]
})

# print(df1)# print(df2)
# df1.join(df2,lsuffix = "_caller",rsuffix = "_other") # default (how = "left") 
df1.join(df2.set_index("key"),on = "key")


# Merge Function
dfm1 = pd.DataFrame({
    "lkey":["A","B","C","F"],
    "value":[1,2,3,4]
})
dfm2 = pd.DataFrame({
    "rkey":["A","B","C","D"],
    "value":[4,5,6,7]
})
dfm1.merge(dfm2,left_on = "lkey",right_on = "rkey",how = "left") # default (how = "inner")

# Concat Function
dfc1 = pd.DataFrame({
    "key":["A","B","C","F"],
    "value1":[1,2,3,4]
})
dfc2 = pd.DataFrame({
    "key":["A","B","C"],
    "value2":[1,2,3]
})
pd.concat([dfc1 , dfc2],ignore_index = True)
```
